### **题目 1：手动实现前馈神经网络解决上述回归、二分类、多分类任务**

#### 1.1  回归问题

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

#=======================加载数据==============================#
# 设置参数
np.random.seed(42)
torch.manual_seed(42)
n_samples=10000
n_features=500

# 生成线性数据
X = np.random.randn(n_samples, n_features)
weight = np.full(n_features, 0.0056)
bias = 0.028
delta = np.random.randn(n_samples) * 0.1
y = bias + X @ weight + delta

# 划分数据集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 将数据转成 PyTorch 张量
X_train = torch.from_numpy(X_train).float()
y_train = torch.from_numpy(y_train).float().reshape(-1, 1)
X_test = torch.from_numpy(X_test).float()
y_test = torch.from_numpy(y_test).float().reshape(-1, 1)
print("--- 数据加载完毕 ---")
#====================数据加载完毕=============================#

# ====================创建模型=================================
class RegerssionModelV0:
    def __init__(self, n_input, n_hidden, n_output):
        self.n_input = n_input
        self.n_hidden = n_hidden
        self.n_output = n_output
        
        self.W1 = torch.randn(n_input, n_hidden) * 0.01
        self.b1 = torch.zeros(1, n_hidden)
        self.W2 = torch.randn(n_hidden, n_output) * 0.01
        self.b2 = torch.zeros(1, n_output)  
        
    def forward(self, X):
        Z1 = X @ self.W1 + self.b1
        self.A1 = torch.relu(Z1)  
        y_pred = self.A1 @ self.W2 + self.b2
        return y_pred   

# 创建实例
input_size = n_features
hidden_size = 256
output_size = 1 
model_0 = RegerssionModelV0(input_size, hidden_size, output_size)
print("--- 模型创建完毕 ---")
#  ===============创建模型完毕===============================#

#=====================训练模型=================================
# 选择损失函数
def mean_squared_error(y_pred, y_true):
    return torch.mean((y_pred - y_true) ** 2)

learning_rate = 0.1
epochs = 400

train_losses = []
test_losses = []
m = X_train.shape[0]

print("--- 开始训练 ---")
for epoch in range(epochs):
    # --- 1. 前向传播：计算预测值 ---
    y_pred = model_0.forward(X_train)
    
    # --- 2. 计算训练损失 ---
    loss = mean_squared_error(y_pred, y_train)
    train_losses.append(loss.item()) 
    
    # --- 3. 反向传播：计算梯度  ---
    d_loss_y_pred = 2 * (y_pred - y_train) / m
    
    # 梯度: 输出层 -> 隐藏层 (W2, b2)
    d_W2 = model_0.A1.T @ d_loss_y_pred
    d_b2 = torch.sum(d_loss_y_pred, dim=0, keepdim=True)
    
    # 梯度: 隐藏层 -> 输入层 (W1, b1)
    d_A1 = d_loss_y_pred @ model_0.W2.T
    d_relu = (model_0.A1 > 0).float()
    d_Z1 = d_A1 * d_relu
    d_W1 = X_train.T @ d_Z1
    d_b1 = torch.sum(d_Z1, dim=0, keepdim=True)
    
    # --- 4. 更新参数 ---
    model_0.W1 -= learning_rate * d_W1
    model_0.b1 -= learning_rate * d_b1
    model_0.W2 -= learning_rate * d_W2
    model_0.b2 -= learning_rate * d_b2
    
    # --- 在测试集上评估性能 ---
    with torch.no_grad():
        test_pred = model_0.forward(X_test)
        test_loss = mean_squared_error(test_pred, y_test)
        test_losses.append(test_loss.item())

    if (epoch + 1) % 100 == 0:
        print(f"Epoch [{epoch+1}/{epochs}], 训练损失: {loss.item():.4f}, 测试损失: {test_loss.item():.4f}")

print("--- 训练完成 ---")
# ====================训练模型完毕============================#

#=================模型预测与可视化=============================#
print("\n--- 模型预测与评估 ---")
with torch.no_grad(): 
    final_test_pred = model_0.forward(X_test)

# 计算并打印 R² 分数
r2 = r2_score(y_test.numpy(), final_test_pred.numpy())
print(f"在测试集上的 R²分数: {r2:.4f}")

print("\n预测值 vs 真实值 (前5个样本):")
for i in range(5):
    print(f"  样本 {i+1}: 预测值 = {final_test_pred[i].item():.4f}, 真实值 = {y_test[i].item():.4f}")

# 可视化
plt.figure(figsize=(12, 5))
plt.plot(train_losses, label='Train Loss')
plt.plot(test_losses, label='Test Loss')
plt.title('损失曲线(MSE)')
plt.xlabel('Epoch')
plt.ylabel('MSE Loss')
plt.legend()
plt.grid(True)
```

**结果分析**
1. **R² 分数**：0.5518，表明模型已经能够解释测试数据中超过55%的方差和规律。
2. **损失曲线分析**：在训练的初期，训练损失与测试损失曲线同步快速下降，在训练的250-300个epochs后进入平稳期，同时在第400个epoch结束时，曲线变得非常平稳，不存在过拟合与欠拟合问题，模型表现良好。

#### 1.2  二分类问题

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

#=======================加载数据==============================#
# 设置参数
np.random.seed(42)
torch.manual_seed(42)
n_samples = 10000  
n_features = 200   

# 生成数据集
center1 = np.ones(n_features) * 0.5
X1, y1 = make_blobs(n_samples=n_samples, centers=[center1], n_features=n_features, cluster_std=1)
y1[:] = 0  # 所有标签设为0

center2 = np.ones(n_features) * -0.5
X2, y2 = make_blobs(n_samples=n_samples, centers=[center2], n_features=n_features, cluster_std=1)
y2[:] = 1  # 所有标签设为1

X = np.vstack([X1, X2])  
y = np.hstack([y1, y2])

# 划分数据集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=42)

# 将数据转化为 PyTorch 张量
X_train = torch.from_numpy(X_train).float()
X_test = torch.from_numpy(X_test).float()
y_train = torch.from_numpy(y_train).float().reshape(-1, 1)
y_test = torch.from_numpy(y_test).float().reshape(-1, 1)

#=======================数据加载完毕===========================#

# ====================创建模型=================================
class ClassificationModelV0:
    def __init__(self, n_input, n_hidden, n_output):
        self.n_input = n_input
        self.n_hidden = n_hidden
        self.n_output = n_output
        
        self.W1 = torch.randn(n_input, n_hidden) * np.sqrt(1. / n_input)
        self.b1 = torch.zeros(1, n_hidden)
        self.W2 = torch.randn(n_hidden, n_output) * np.sqrt(1. / n_hidden)
        self.b2 = torch.zeros(1, n_output)
        
    def forward(self, X):
        # 隐藏层
        Z1 = X @ self.W1 + self.b1
        self.A1 = torch.relu(Z1)
        
        # 输出层
        Z2 = self.A1 @ self.W2 + self.b2
        y_pred = torch.sigmoid(Z2) # 使用 Sigmoid 激活函数处理二分类问题
        return y_pred

# 创建实例
input_size = n_features
hidden_size = 128 
output_size = 1 
model_1 = ClassificationModelV0(input_size, hidden_size, output_size)
print("--- 模型创建完毕 ---")
#  ===============创建模型完毕===============================#

#=====================训练模型=================================
# 选择损失函数 (二元交叉熵)
def binary_cross_entropy(y_pred, y_true):
    epsilon = 1e-12
    y_pred = torch.clamp(y_pred, epsilon, 1 - epsilon)
    return -torch.mean(y_true * torch.log(y_pred) + (1 - y_true) * torch.log(1 - y_pred))

# 设置超参数
learning_rate = 0.05
epochs = 200

train_losses = []
test_losses = []
m = X_train.shape[0]

print("--- 开始训练 ---")
for epoch in range(epochs):
    # --- 1. 前向传播：计算预测值 ---
    y_pred = model_1.forward(X_train)
    
    # --- 2. 计算训练损失 ---
    loss = binary_cross_entropy(y_pred, y_train)
    train_losses.append(loss.item())
    
    # --- 3. 反向传播：计算梯度 ---
    # 输出层的梯度 (d(Loss)/d(y_pred))
    d_loss_y_pred = - (y_train / y_pred) + ((1 - y_train) / (1 - y_pred))
    d_loss_y_pred /= m
    
    # Sigmoid的梯度 (d(y_pred)/d(Z2))
    d_sigmoid = y_pred * (1 - y_pred)
    d_loss_Z2 = d_loss_y_pred * d_sigmoid
    
    # 梯度: 输出层 -> 隐藏层 (W2, b2)
    d_W2 = model_1.A1.T @ d_loss_Z2
    d_b2 = torch.sum(d_loss_Z2, dim=0, keepdim=True)
    
    # 梯度: 隐藏层 -> 输入层 (W1, b1)
    d_A1 = d_loss_Z2 @ model_1.W2.T
    d_relu = (model_1.A1 > 0).float()
    d_Z1 = d_A1 * d_relu
    d_W1 = X_train.T @ d_Z1
    d_b1 = torch.sum(d_Z1, dim=0, keepdim=True)
    
    # --- 4. 更新参数 ---
    model_1.W1 -= learning_rate * d_W1
    model_1.b1 -= learning_rate * d_b1
    model_1.W2 -= learning_rate * d_W2
    model_1.b2 -= learning_rate * d_b2
    
    # --- 在测试集上评估性能 ---
    with torch.no_grad():
        test_pred_probs = model_1.forward(X_test)
        test_loss = binary_cross_entropy(test_pred_probs, y_test)
        test_losses.append(test_loss.item())
        
        # 将概率转换为类别 (0 或 1)
        test_pred_labels = (test_pred_probs > 0.5).float()
        accuracy = accuracy_score(y_test.numpy(), test_pred_labels.numpy())

    if (epoch + 1) % 20 == 0:
        print(f"Epoch [{epoch+1}/{epochs}], 训练损失: {loss.item():.4f}, 测试损失: {test_loss.item():.4f}, 测试准确率: {accuracy:.4f}")

print("--- 训练完成 ---")
# ====================训练模型完毕============================#


#=================模型预测与可视化=============================#
print("\n--- 模型预测与评估 ---")
with torch.no_grad(): 
    final_test_pred_probs = model_1.forward(X_test)
    final_test_pred_labels = (final_test_pred_probs > 0.5).float()

# 计算并打印准确率
final_accuracy = accuracy_score(y_test.numpy(), final_test_pred_labels.numpy())
print(f"在测试集上的准确率: {final_accuracy:.4f}")

print("\n预测概率 vs 真实标签 (前10个样本):")
for i in range(10):
    print(f"  样本 {i+1}: 预测概率 = {final_test_pred_probs[i].item():.4f} (预测类别: {final_test_pred_labels[i].item():.0f}), 真实标签 = {y_test[i].item():.0f}")

# 可视化损失曲线
plt.figure(figsize=(12, 5))
plt.plot(train_losses, label='Train Loss')
plt.plot(test_losses, label='Test Loss')
plt.title('损失曲线 (二元交叉熵)')
plt.xlabel('Epoch')
plt.ylabel('BCE Loss')
plt.legend()
plt.grid(True)
plt.show()
```

**结果分析**
1. **准确率**：1.0000，表明模型在测试集上达到了100%的分类准确率，能够完美地区分两个类别的数据。
2. **损失曲线分析**：在训练初期（约前25个epochs），两条曲线同步快速下降。随后，损失平稳地持续降低，最终在200个epoch结束时收敛到一个非常低的值，不存在过拟合与欠拟合问题，模型表现良好。

#### 1.3  多分类问题

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import torchvision
import torchvision.transforms as transforms
from sklearn.metrics import accuracy_score

#=======================加载数据==============================#
# 加载MNIST数据集
train_dataset = torchvision.datasets.MNIST(root='~/Datasets/MNIST', train=True, download=True, transform=transforms.ToTensor())
test_dataset = torchvision.datasets.MNIST(root='~/Datasets/MNIST', train=False, download=True, transform=transforms.ToTensor())

# 通过DataLoader读取小批量数据样本
BATCH_SIZE = 64 # 定义批次大小
train_loader = torch.utils.data.DataLoader(dataset=train_dataset, batch_size=BATCH_SIZE, shuffle=True)
test_loader = torch.utils.data.DataLoader(dataset=test_dataset, batch_size=BATCH_SIZE, shuffle=False)

print(f"成功创建 DataLoader, 每个批次大小为: {BATCH_SIZE}")
#=======================数据加载完毕===========================#


# ====================创建模型=================================
class MultiClassModelV0:
    def __init__(self, n_input, n_hidden, n_output):
        self.n_input = n_input
        self.n_hidden = n_hidden
        self.n_output = n_output
        self.W1 = torch.randn(n_input, n_hidden) * np.sqrt(1. / n_input)
        self.b1 = torch.zeros(1, n_hidden)
        self.W2 = torch.randn(n_hidden, n_output) * np.sqrt(1. / n_hidden)
        self.b2 = torch.zeros(1, n_output)
        
    def forward(self, X):
        Z1 = X @ self.W1 + self.b1
        self.A1 = torch.relu(Z1)
        Z2 = self.A1 @ self.W2 + self.b2
        exp_Z2 = torch.exp(Z2 - torch.max(Z2, dim=1, keepdim=True).values)
        y_pred = exp_Z2 / torch.sum(exp_Z2, dim=1, keepdim=True)
        return y_pred

# 创建实例
input_size = 784
hidden_size = 128
output_size = 10 
model_2 = MultiClassModelV0(input_size, hidden_size, output_size)
print("\n--- 模型创建完毕 ---")
#  ===============创建模型完毕===============================#


#=====================训练模型=================================
def cross_entropy_loss(y_pred, y_true_one_hot):
    epsilon = 1e-12
    y_pred = torch.clamp(y_pred, epsilon, 1. - epsilon)
    loss = -torch.sum(y_true_one_hot * torch.log(y_pred), dim=1)
    return torch.mean(loss)

# 设置超参数
learning_rate = 0.05
epochs = 15
epoch_train_losses = []
epoch_test_losses = []

print("--- 开始小批量训练 ---")
for epoch in range(epochs):
    batch_train_loss = 0.0
    for X_batch, y_batch in train_loader:
        # 1. 展平图像数据
        X_batch_flat = X_batch.view(X_batch.shape[0], -1)
        # 2. 将标签转换为 One-Hot 编码
        y_batch_one_hot = torch.nn.functional.one_hot(y_batch, num_classes=10).float()
        
        # --- 1. 前向传播 (在小批次上) ---
        y_pred = model_2.forward(X_batch_flat)
        
        # --- 2. 计算当前批次的损失 ---
        loss = cross_entropy_loss(y_pred, y_batch_one_hot)
        
        # --- 3. 反向传播：计算梯度 ---
        m_batch = X_batch_flat.shape[0] # 注意：m 现在是小批次的样本数
        d_loss_Z2 = (y_pred - y_batch_one_hot) / m_batch
        
        d_W2 = model_2.A1.T @ d_loss_Z2
        d_b2 = torch.sum(d_loss_Z2, dim=0, keepdim=True)
        
        d_A1 = d_loss_Z2 @ model_2.W2.T
        d_relu = (model_2.A1 > 0).float()
        d_Z1 = d_A1 * d_relu
        
        d_W1 = X_batch_flat.T @ d_Z1
        d_b1 = torch.sum(d_Z1, dim=0, keepdim=True)
        
        # --- 4. 更新参数 (每个批次都更新一次) ---
        model_2.W1 -= learning_rate * d_W1
        model_2.b1 -= learning_rate * d_b1
        model_2.W2 -= learning_rate * d_W2
        model_2.b2 -= learning_rate * d_b2

        # 累加当前批次的损失
        batch_train_loss += loss.item()
        
    # --- 计算并记录整个epoch的平均损失 ---
    avg_train_loss = batch_train_loss / len(train_loader)
    epoch_train_losses.append(avg_train_loss)
    
    # --- 在测试集上评估性能 ---
    test_loss = 0.0
    test_acc = 0.0
    with torch.no_grad():
        for X_test_batch, y_test_batch in test_loader:
            X_test_batch_flat = X_test_batch.view(X_test_batch.shape[0], -1)
            y_test_batch_one_hot = torch.nn.functional.one_hot(y_test_batch, num_classes=10).float()
            
            test_pred_probs = model_2.forward(X_test_batch_flat)
            test_loss += cross_entropy_loss(test_pred_probs, y_test_batch_one_hot).item()
            
            test_pred_labels = torch.argmax(test_pred_probs, dim=1)
            test_acc += accuracy_score(y_test_batch.numpy(), test_pred_labels.numpy())

    avg_test_loss = test_loss / len(test_loader)
    avg_test_acc = test_acc / len(test_loader)
    epoch_test_losses.append(avg_test_loss)

    print(f"Epoch [{epoch+1}/{epochs}], 训练损失: {avg_train_loss:.4f}, 测试损失: {avg_test_loss:.4f}, 测试准确率: {avg_test_acc:.4f}")

print("--- 训练完成 ---")
# ====================训练模型完毕============================#


#=================模型预测与可视化=============================#
# 可视化损失曲线
plt.figure(figsize=(12, 5))
plt.plot(epoch_train_losses, label='Train Loss')
plt.plot(epoch_test_losses, label='Test Loss')
plt.title('损失曲线 (交叉熵 - 小批量训练)')
plt.xlabel('Epoch')
plt.ylabel('Cross-Entropy Loss')
plt.legend()
plt.grid(True)
plt.show()
```

**结果分析**
1. **准确率**：模型在15个周期结束时，在测试集上的最终准确率为 97.32%，说明模型对绝大多数从未见过的数据都能进行正确的分类。
2. **损失曲线分析**：训练损失和测试损失都在持续、平滑地下降。不存在过拟合与欠拟合问题，模型表现良好。

### **题目 2：利用torch.nn实现前馈神经网络解决上述回归、二分类、多分类任务**

#### 2.1  回归问题

```python
import torch
import torch.nn as nn 
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import numpy as np
import matplotlib.pyplot as plt

#=======================加载数据==============================#
np.random.seed(42)
torch.manual_seed(42)
n_samples=10000
n_features=500
X = np.random.randn(n_samples, n_features)
weight = np.full(n_features, 0.0056)
bias = 0.028
delta = np.random.randn(n_samples) * 0.1
y = bias + X @ weight + delta
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_train = torch.from_numpy(X_train).float()
y_train = torch.from_numpy(y_train).float().reshape(-1, 1)
X_test = torch.from_numpy(X_test).float()
y_test = torch.from_numpy(y_test).float().reshape(-1, 1)
print("--- 数据加载完毕 ---")
#====================数据加载完毕=============================#

# ====================创建模型================================
# 定义模型超参数
input_size = n_features
hidden_size = 256
output_size = 1 

model_3 = nn.Sequential(
    nn.Linear(input_size, hidden_size), 
    nn.ReLU(), # ReLU激活函数
    nn.Linear(hidden_size, output_size) )

print("--- 模型创建完毕 ---")
print("模型结构:")
print(model_3)
#  ===============创建模型完毕===============================#

#=====================训练模型=================================
# 定义损失函数和优化器 
loss_fn = nn.MSELoss() 
learning_rate = 0.01
optimizer = torch.optim.Adam(params=model_3.parameters(), lr=learning_rate,weight_decay=1e-4) 

# 定义训练循环 
epochs = 1000
train_losses = []
test_losses = []

print("--- 开始训练 ---")
for epoch in range(epochs):
    ### 训练阶段
    model_3.train() 
    
    # 1. 前向传播
    y_pred = model_3(X_train)
    
    # 2. 计算损失
    loss = loss_fn(y_pred, y_train)
    train_losses.append(loss.item())
    
    # 3. 反向传播和优化 
    optimizer.zero_grad() 
    loss.backward()       
    optimizer.step()     

    ### 测试阶段
    model_3.eval() 
    with torch.no_grad(): 
        test_pred = model_3(X_test)
        test_loss = loss_fn(test_pred, y_test)
        test_losses.append(test_loss.item())

    if (epoch + 1) % 200 == 0:
        print(f"Epoch [{epoch+1}/{epochs}], 训练损失: {loss.item():.4f}, 测试损失: {test_loss.item():.4f}")

print("--- 训练完成 ---")
# ====================训练模型完毕============================#

#=================模型预测与可视化=============================#
print("\n--- 模型预测与评估 ---")
model_3.eval() 
with torch.no_grad(): 
    final_test_pred = model_3(X_test)

# 计算并打印 R² 分数
r2 = r2_score(y_test.numpy(), final_test_pred.numpy())
print(f"在测试集上的 R²分数: {r2:.4f}")

print("\n预测值 vs 真实值 (前5个样本):")
for i in range(5):
    print(f"  样本 {i+1}: 预测值 = {final_test_pred[i].item():.4f}, 真实值 = {y_test[i].item():.4f}")

# 可视化
plt.figure(figsize=(12, 5))
plt.plot(train_losses, label='Train Loss')
plt.plot(test_losses, label='Test Loss')
plt.title('损失曲线 ')
plt.xlabel('Epoch')
plt.ylabel('MSE Loss')
plt.legend()
plt.grid(True)
plt.show()
```

**结果分析**
1. **R² 分数**：0.5520，表明模型已经能够解释测试数据中超过55.2%的方差和规律,证明了Adam优化器的有效性。
2. **损失曲线分析**：在训练的初期，训练损失与测试损失曲线都以极快的速度下降到一个非常低的水平，在训练的100-400个epochs后进入平稳期，趋近于零。同时在第1000个epoch结束时，训练损失持续保持在零附近，不存在过拟合与欠拟合问题，模型表现良好。

#### 2.2  二分类问题

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from torch import nn

#=======================加载数据==============================#
# 设置参数
np.random.seed(42)
torch.manual_seed(42)
n_samples = 10000  
n_features = 200   

# 生成数据集
center1 = np.ones(n_features) * 0.5
X1, y1 = make_blobs(n_samples=n_samples, centers=[center1], n_features=n_features, cluster_std=1)
y1[:] = 0  # 所有标签设为0

center2 = np.ones(n_features) * -0.5
X2, y2 = make_blobs(n_samples=n_samples, centers=[center2], n_features=n_features, cluster_std=1)
y2[:] = 1  # 所有标签设为1

X = np.vstack([X1, X2])  
y = np.hstack([y1, y2])

# 划分数据集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=42)

# 将数据转化为 PyTorch 张量
X_train = torch.from_numpy(X_train).float()
X_test = torch.from_numpy(X_test).float()
y_train = torch.from_numpy(y_train).float().reshape(-1, 1)
y_test = torch.from_numpy(y_test).float().reshape(-1, 1)

#=======================数据加载完毕===========================#

# ====================创建模型================================
from torch import nn
class ClassificationModelV0(nn.Module):
    def __init__(self):
        super().__init__()
        self.layers = nn.Sequential(
            nn.Linear(in_features=200, out_features=168),
            nn.ReLU(),
            nn.Linear(in_features=168, out_features=1)
        )

    def forward(self, x):
       return self.layers(x)

model_4 = ClassificationModelV0()
print("--- 模型创建完毕 ---")
print("模型结构:")
print(model_4)
#  ===============创建模型完毕===============================#

#=====================训练模型=================================
# 定义损失函数和优化器 
loss_fn = nn.BCEWithLogitsLoss() #
learning_rate = 0.01
optimizer = torch.optim.Adam(params=model_4.parameters(), lr=learning_rate,weight_decay=1e-4) 

# 定义训练循环 
epochs = 50
train_losses = []
test_losses = []
train_accuracies = []
test_accuracies = []

print("--- 开始训练 ---")
for epoch in range(epochs):
    ### 训练阶段
    model_4.train() 
    
    # 1. 前向传播
    y_logits = model_4(X_train) 
    
    # 2. 计算损失
    loss = loss_fn(y_logits, y_train)
    train_losses.append(loss.item()) 
    y_pred_labels = torch.round(torch.sigmoid(y_logits))
    train_acc = accuracy_score(y_true=y_train.numpy(), y_pred=y_pred_labels.detach().numpy())
    train_accuracies.append(train_acc)

    # 3. 反向传播和优化 
    optimizer.zero_grad() 
    loss.backward()       
    optimizer.step()     

    ### 测试阶段
    model_4.eval() 
    with torch.no_grad(): 
        test_logits = model_4(X_test)
        test_loss = loss_fn(test_logits, y_test)
        test_losses.append(test_loss.item()) 
        test_pred_labels = torch.round(torch.sigmoid(test_logits))
        test_acc = accuracy_score(y_true=y_test.numpy(), y_pred=test_pred_labels.numpy())
        test_accuracies.append(test_acc)

    if epoch % 10 == 0:
        print(f"Epoch: {epoch} | Train Loss: {loss:.5f}, Train Acc: {train_acc:.2f}% | Test Loss: {test_loss:.5f}, Test Acc: {test_acc:.2f}%")

print("--- 训练完成 ---")
# ====================训练模型完毕============================#

#=================模型预测与可视化=============================#
print("\n--- 模型预测与评估 ---")
model_4.eval() 
with torch.no_grad(): 
    final_test_logits = model_4(X_test)
    final_test_pred_labels = torch.round(torch.sigmoid(final_test_logits))
final_accuracy = accuracy_score(y_test.numpy(), final_test_pred_labels.numpy())
print(f"在测试集上的最终准确率: {final_accuracy:.4f}")

print("\n预测类别 vs 真实类别 (前10个样本):")
for i in range(10):
    print(f"  样本 {i+1}: 预测类别 = {final_test_pred_labels[i].item():.0f}, 真实类别 = {y_test[i].item():.0f}")

# 可视化
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(train_losses, label='Train Loss')
plt.plot(test_losses, label='Test Loss')
plt.title('损失曲线')
plt.xlabel('Epoch')
plt.ylabel('BCEWithLogits Loss')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(train_accuracies, label='Train Accuracy')
plt.plot(test_accuracies, label='Test Accuracy')
plt.title('准确率曲线')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import torchvision
import torchvision.transforms as transforms
from torch import nn 
from torch.utils.data import DataLoader
from sklearn.metrics import accuracy_score

#======================= 加载数据==========================#

BATCH_SIZE = 64
train_dataset = torchvision.datasets.MNIST(root='~/Datasets/MNIST', train=True, download=True, transform=transforms.ToTensor())
test_dataset = torchvision.datasets.MNIST(root='~/Datasets/MNIST', train=False, download=True, transform=transforms.ToTensor())

train_loader = DataLoader(dataset=train_dataset, batch_size=BATCH_SIZE, shuffle=True)
test_loader = DataLoader(dataset=test_dataset, batch_size=BATCH_SIZE, shuffle=False)

print(f"数据加载完毕，每个批次大小为: {BATCH_SIZE}")
#=======================数据加载完毕===========================#


# ==================== 创建模型=========================
class MultiClassModelV0(nn.Module):
    def __init__(self, input_features, output_features, hidden_units=128):
        super().__init__()
        self.layers = nn.Sequential(
            nn.Flatten(), 
            nn.Linear(in_features=input_features, out_features=hidden_units),
            nn.ReLU(),
            nn.Linear(in_features=hidden_units, out_features=output_features)
        )

    def forward(self, x):
        return self.layers(x)

model_5 = MultiClassModelV0(input_features=784, output_features=10)
print("\n--- 模型创建完毕 ---")
print(model_5)
#  ===============创建模型完毕===============================#

#定义损失函数和优化器 
loss_fn = nn.CrossEntropyLoss()

learning_rate = 0.001
optimizer = torch.optim.SGD(params=model_5.parameters(), lr=learning_rate)


#=================== 训练模型 =================================
epochs = 100
epoch_train_losses, epoch_test_losses = [], []
epoch_train_accs, epoch_test_accs = [], []

print("\n--- 开始训练---")
for epoch in range(epochs):
    # --- 训练阶段 ---
    model_5.train() # 将模型设置为训练模式
    train_loss, train_acc = 0, 0
    for X_batch, y_batch in train_loader:
        # 1. 前向传播 
        y_logits = model_5(X_batch)
        
        # 2. 计算损失
        loss = loss_fn(y_logits, y_batch)
        train_loss += loss.item()

        # 计算准确率
        y_pred_class = torch.argmax(torch.softmax(y_logits, dim=1), dim=1)
        train_acc += accuracy_score(y_batch.numpy(), y_pred_class.numpy())

        # 3. 优化器清零梯度
        optimizer.zero_grad()
        
        # 4. 反向传播 (PyTorch 自动计算所有梯度)
        loss.backward()
        
        # 5. 参数更新 (优化器根据梯度更新所有参数)
        optimizer.step()
        
    # 计算每个 epoch 的平均损失和准确率
    train_loss /= len(train_loader)
    train_acc /= len(train_loader)
    epoch_train_losses.append(train_loss)
    epoch_train_accs.append(train_acc)

    # --- 测试阶段 ---
    model_5.eval() # 将模型设置为评估模式
    test_loss, test_acc = 0, 0
    with torch.no_grad(): # 在评估时关闭梯度计算，节省资源
        for X_test_batch, y_test_batch in test_loader:
            test_logits = model_5(X_test_batch)
            test_loss += loss_fn(test_logits, y_test_batch).item()
            
            test_pred_class = torch.argmax(torch.softmax(test_logits, dim=1), dim=1)
            test_acc += accuracy_score(y_test_batch.numpy(), test_pred_class.numpy())
            
    test_loss /= len(test_loader)
    test_acc /= len(test_loader)
    epoch_test_losses.append(test_loss)
    epoch_test_accs.append(test_acc)
    if (epoch + 1) % 20 == 0:
        print(f"Epoch [{epoch+1}/{epochs}], Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.4f} | Test Loss: {test_loss:.4f}, Test Acc: {test_acc:.4f}")

print("--- 训练完成 ---")
# ====================训练模型完毕============================#


#================= 模型预测与可视化 ==========================
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(epoch_train_losses, label='Train Loss')
plt.plot(epoch_test_losses, label='Test Loss')
plt.title('损失曲线 (nn.Module)')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(epoch_train_accs, label='Train Accuracy')
plt.plot(epoch_test_accs, label='Test Accuracy')
plt.title('准确率曲线 (nn.Module)')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

### **题目 4：在多分类任务实验中手动实现一种解决过拟合的方法（$L_2$正则化或dropout）**

* 探究使用前后模型效果的变化，并完成下述任务中的一项
* $L_2$正则化：探究惩罚项的权重对实验结果的影响（可用loss曲线进行展示）
* dropout：探究不同丢弃率对实验结果的影响（可用loss曲线进行展示）

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import torchvision
import torchvision.transforms as transforms
from torch import nn
from torch.utils.data import DataLoader
from sklearn.metrics import accuracy_score

#======================= 加载数据 ==============================#
# 使用 DataLoader 进行小批量训练，这是更标准和高效的做法
BATCH_SIZE = 64
train_dataset = torchvision.datasets.MNIST(root='~/Datasets/MNIST', train=True, download=True, transform=transforms.ToTensor())
test_dataset = torchvision.datasets.MNIST(root='~/Datasets/MNIST', train=False, download=True, transform=transforms.ToTensor())

train_loader = DataLoader(dataset=train_dataset, batch_size=BATCH_SIZE, shuffle=True)
test_loader = DataLoader(dataset=test_dataset, batch_size=BATCH_SIZE, shuffle=False)
print("数据加载完毕。")
#=======================数据加载完毕===========================#


# ==================== 创建模型 (使用 nn.Module) =================
class MNISTModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.layers = nn.Sequential(
            nn.Flatten(),
            nn.Linear(in_features=784, out_features=128),
            nn.ReLU(),
            nn.Linear(in_features=128, out_features=10)
        )

    def forward(self, x):
        return self.layers(x)

# ==================== 训练与对比 =============================
# 定义不同 L2 正则化权重 
lambda_values = [0, 0.001, 0.01, 0.1] 
results = {}

# 设置训练超参数
learning_rate = 0.05
epochs = 50

print("\n--- 开始对比不同L2正则化权重的效果 (使用PyTorch优化器) ---")

# 遍历每一种 lambda 值
for lambda_reg in lambda_values:
    print(f"\n===== 正在使用 weight_decay = {lambda_reg} 进行训练 =====")
    
    # 1. 为每个实验创建一个全新的模型实例
    model = MNISTModel()
    
    # 2. 定义损失函数 (不再需要手动添加L2惩罚项)
    loss_fn = nn.CrossEntropyLoss()
    
    # 3. 定义优化器 (核心改动点：在这里传入 weight_decay)
    # weight_decay 参数就是PyTorch中实现L2正则化的方式
    optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate, weight_decay=lambda_reg)
    
    # 4. 准备记录这个模型的结果
    test_losses = []

    # 5. 训练循环 (使用小批量)
    for epoch in range(epochs):
        model.train()
        for X_batch, y_batch in train_loader:
            y_logits = model(X_batch)
            loss = loss_fn(y_logits, y_batch)
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

        # --- 在测试集上评估性能 ---
        model.eval()
        current_test_loss = 0
        with torch.no_grad():
            for X_test_batch, y_test_batch in test_loader:
                test_logits = model(X_test_batch)
                # 注意：这里的损失只包含交叉熵，是评估模型泛化能力的干净指标
                test_loss = loss_fn(test_logits, y_test_batch)
                current_test_loss += test_loss.item()
        
        avg_test_loss = current_test_loss / len(test_loader)
        test_losses.append(avg_test_loss)

    # 6. 保存这个lambda值的所有训练历史
    results[lambda_reg] = {"test_loss": test_losses}
    print(f"--- weight_decay = {lambda_reg} 训练完成 ---")


# ====================  结果可视化与分析 =======================
print("\n--- 正在绘制对比图表 ---")
plt.figure(figsize=(12, 8))

for lambda_reg, data in results.items():
    plt.plot(data["test_loss"], label=f"Test Loss (weight_decay={lambda_reg})")

plt.title("不同L2正则化权重对测试损失的影响", fontsize=16)
plt.xlabel("Epoch")
plt.ylabel("Cross-Entropy Loss")
plt.legend()
plt.grid(True)
plt.show()
```

**结果分析**
1. **weight_decay=0**(蓝色曲线 - 无正则化)这是我们的基准模型，不施加任何权重惩罚。它的测试损失下降得最快，并且最终收敛到了最低的水平（约0.08）。模型本身并没有出现严重的过拟合，因此不加正则化的效果已经非常好。

2. **weight_decay=0.001** (橙色曲线 - 轻微正则化)这条曲线的表现与无正则化的基准模型非常接近，最终的损失值也几乎相同，略高一点点。这说明一个较小的惩罚权重对模型的影响不大，既没有显著提升性能，也没有造成负面影响。

3. **weight_decay=0.01** (绿色曲线 - 中等正则化)这条曲线的表现明显差于前两者。它的损失值从始至终都处于一个较高的水平，最终收敛在约0.25左右。这种情况我们称之为欠拟合。

4. **weight_decay=0.1** (红色曲线 - 强正则化)这条曲线的表现是四者中最差的。它的初始损失值最高，并且始终远高于其他所有曲线。0.1的惩罚力度是极强的，严重阻碍了模型的学习过程，导致了更严重的欠拟合。


```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import torchvision
import torchvision.transforms as transforms
from sklearn.metrics import accuracy_score

#================== 加载数据 ==============================#
train_dataset = torchvision.datasets.MNIST(root='~/Datasets/MNIST', train=True, download=True, transform=transforms.ToTensor())
test_dataset = torchvision.datasets.MNIST(root='~/Datasets/MNIST', train=False, download=True, transform=transforms.ToTensor())

X_train = train_dataset.data.view(len(train_dataset), -1).float() / 255.0
y_train = train_dataset.targets
X_test = test_dataset.data.view(len(test_dataset), -1).float() / 255.0
y_test = test_dataset.targets

y_train_one_hot = torch.nn.functional.one_hot(y_train, num_classes=10).float()
y_test_one_hot = torch.nn.functional.one_hot(y_test, num_classes=10).float()
print("数据加载完毕。")
#=======================数据加载完毕===========================#


# ================== 创建支持Dropout的模型 =================
class DropoutModel:
    def __init__(self, n_input, n_hidden, n_output, dropout_rate=0.5):
        self.W1 = torch.randn(n_input, n_hidden) * np.sqrt(1. / n_input)
        self.b1 = torch.zeros(1, n_hidden)
        self.W2 = torch.randn(n_hidden, n_output) * np.sqrt(1. / n_hidden)
        self.b2 = torch.zeros(1, n_output)
        
        # --- Dropout 相关参数 ---
        self.p = dropout_rate # 丢弃率
        self.keep_prob = 1 - self.p # 保留率
        
    def forward(self, X, training=True):
        Z1 = X @ self.W1 + self.b1
        A1 = torch.relu(Z1)
        
        # 在前向传播中应用Dropout
        if training:
            # D1 中约有 p% 的元素为0, (1-p)% 的元素为1
            self.D1 = (torch.rand_like(A1) < self.keep_prob).float()
            
            # 将蒙版应用到激活值上，关闭一部分神经元
            A1 = A1 * self.D1
            
            # Inverted Dropout: 将保留下来的激活值进行缩放,保证在测试时，隐藏层输出的期望值与训练时一致
            A1 = A1 / self.keep_prob
        
        self.A1 = A1 # 保存A1以用于反向传播
        
        Z2 = self.A1 @ self.W2 + self.b2
        exp_Z2 = torch.exp(Z2 - torch.max(Z2, dim=1, keepdim=True).values)
        return exp_Z2 / torch.sum(exp_Z2, dim=1, keepdim=True)

# ================  训练与对比 =============================
# 定义要对比的不同丢弃率 (p)
# p = 0 表示没有dropout，作为基准对比
dropout_rates = [0, 0.2, 0.5] 

results = {}

# 设置训练超参数
learning_rate = 0.05
epochs = 50
m = X_train.shape[0]

print("\n--- 开始对比不同Dropout率的效果 ---")

for p in dropout_rates:
    print(f"\n===== 正在使用 dropout_rate = {p} 进行训练 =====")
    
    # 1. 为每个实验创建一个全新的模型实例
    model = DropoutModel(n_input=784, n_hidden=128, n_output=10, dropout_rate=p)
    
    test_losses = []

    # 2. 训练循环
    for epoch in range(epochs):
        # --- 前向传播 ---
        y_pred = model.forward(X_train, training=True)
        
        # --- 计算损失 ---
        cross_entropy = -torch.sum(y_train_one_hot * torch.log(torch.clamp(y_pred, 1e-12, 1.0)), dim=1)
        loss = torch.mean(cross_entropy)
        
        # --- 反向传播 ---
        d_loss_Z2 = (y_pred - y_train_one_hot) / m
        
        d_W2 = model.A1.T @ d_loss_Z2
        d_b2 = torch.sum(d_loss_Z2, dim=0, keepdim=True)
        
        d_A1 = d_loss_Z2 @ model.W2.T
        d_relu = (model.A1 > 0).float()
        
        d_Z1 = d_A1 * d_relu
        if p > 0: # 只有在启用dropout时才需要处理
             d_Z1 = d_Z1 * model.D1 / model.keep_prob

        d_W1 = X_train.T @ d_Z1
        d_b1 = torch.sum(d_Z1, dim=0, keepdim=True)
        
        # --- 更新参数 ---
        model.W1 -= learning_rate * d_W1
        model.b1 -= learning_rate * d_b1
        model.W2 -= learning_rate * d_W2
        model.b2 -= learning_rate * d_b2

        # --- 在测试集上评估性能  ---
        with torch.no_grad():
            # 测试时，training=False，不使用dropout
            test_pred_probs = model.forward(X_test, training=False)
            test_ce_loss = -torch.sum(y_test_one_hot * torch.log(torch.clamp(test_pred_probs, 1e-12, 1.0)), dim=1)
            final_test_loss = torch.mean(test_ce_loss)
            test_losses.append(final_test_loss.item())

    results[p] = {"test_loss": test_losses}
    print(f"--- dropout_rate = {p} 训练完成 ---")


# ================== 结果可视化与分析 =======================
print("\n--- 正在绘制对比图表 ---")
plt.figure(figsize=(12, 8))

for p, data in results.items():
    plt.plot(data["test_loss"], label=f"Test Loss (dropout_rate={p})")

plt.title("不同Dropout率对测试损失的影响", fontsize=16)
plt.xlabel("Epoch")
plt.ylabel("Cross-Entropy Loss")
plt.legend()
plt.grid(True)
plt.show()
```

**结果分析**
1. **dropout_rate=0** (蓝色曲线 - 无Dropout)这是我们的基准模型，不进行任何正则化。在这50个周期内，它的损失下降速度最快，并且最终达到了最低的损失值。这表明在没有正则化的情况下，模型能够最快地学习数据中的模式。

2. **dropout_rate=0.2** (橙色曲线 - 轻微Dropout)这条曲线的表现与基准模型非常接近，但全程的损失值都略高于基准模型。这说明，加入轻微的Dropout后，模型的学习速度受到了一点影响，收敛得比基准模型稍慢。

3. **dropout_rate=0.5** (绿色曲线 - 较强Dropout)这条曲线的表现是三者中最差的。它的损失值始终远高于另外两条曲线，收敛速度最慢。0.5的丢弃率是一个非常强的正则化约束。在训练的每一步都随机“关闭”一半的神经元，这严重减缓了模型的学习效率。

## 五、实验心得体会
&emsp;&emsp;通过本次实验，我不仅系统性地应用了前馈神经网络理论知识，更通过亲手实践与框架应用两种方式，对神经网络的工作原理有了更具体、更深入的理解。

&emsp;&emsp;通过手动实现反向传播，对比回归任务的均方误差（`MSE`）求导，与二分类与多分类任务中基于交叉熵损失的梯度计算，我明白了梯度是如何从损失函数逐层传递，并最终用于更新参数的。相比之下，PyTorch框架通过`loss.backward()`和`optimizer.step()`实现所有梯度计算和参数更新，更为便捷快速。

&emsp;&emsp;在多分类任务中，`ReLU`、`Tanh`和`Sigmoid`三种激活函数的对比实验结果非常鲜明。`ReLU`函数在收敛速度和最终准确率上都表现最佳。在本次实验中，由于模型本身并未出现严重的过拟合现象，引入较强的正则化（如高`weight_decay`值或高`dropout_rate`）反而限制了模型的学习能力，导致了欠拟合，测试损失和准确率均不如基准模型。
